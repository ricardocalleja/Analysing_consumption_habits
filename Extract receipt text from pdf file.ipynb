### Extract text with PDFminer

```python
# !pip install pdfminer.six
from io import StringIO

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser

output_string = StringIO()

with open(r'C:\Users\rc_sh\OneDrive\Escritorio\AI\Machine_Learning\Walmart\2020\12_Diciembre\28-12-2020.pdf', 'rb') as in_file:
# with open(r'C:\Users\rc_sh\OneDrive\Escritorio\AI\Machine_Learning\Walmart\2020\Receipts_2020.pdf', 'rb') as in_file:
    parser = PDFParser(in_file)
    doc = PDFDocument(parser)
    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, output_string, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    for page in PDFPage.create_pages(doc):
        interpreter.process_page(page)

print(output_string.getvalue())
```

### Extract fields with Regular Expressions

Regex1: CODIGO:\s(.+?)\n(.+?)[Xx]\s(.+?)\s(.*?)\n

Regex2: BOLETA ELECTRÓNICA Nº :\n\n(.+?)\n\nLOCAL:\n\n(.+?)\n\nCAJA:\n\n(.+?)\n\nFECHA EMISION:\n\n(.+?)\n\nTRAN. Nº:\n\n(.+?)\n\nCAJERO:\n\n(.+?)\n\nHORA:\n\n(.+?)\n\n

Regex3: TOTAL ARTICULOS =(.+?)\n

Regex4: ATENDIDO POR :\s(.+?)\n

```python
import re
# Extract: Code, Quantity, Unit Price, Description
raw_data1 = re.findall('CODIGO:\s(.+?)\n(.+?)[Xx]\s(.+?)\s(.*?)\n' ,  output_string.getvalue(), re.M)
# raw_data1
```

```python
import pandas as pd
df1 = pd.DataFrame(raw_data1, columns =['codigo', 'cantidad', 'precio_unitario', 'descripcion'])
df1
```

```python
raw_data2 = re.findall('BOLETA ELECTRÓNICA Nº :\n\n(.+?)\n\nLOCAL:\n\n(.+?)\n\nCAJA:\n\n(.+?)\n\nFECHA EMISION:\n\n(.+?)\n\nTRAN. Nº:\n\n(.+?)\n\nCAJERO:\n\n(.+?)\n\nHORA:\n\n(.+?)\n\n' ,  output_string.getvalue(), re.M)
# raw_data2
```

```python
df2 = pd.DataFrame(raw_data2, columns =['boleta_electronica', 'local', 'caja', 'fecha_emision', 'trans_n', 'cajero', 'hora']
)
df2
```

```python
raw_data3 = re.findall('TOTAL ARTICULOS =(.+?)\n' ,  output_string.getvalue(), re.M)
# raw_data3
```

```python
df3 = pd.DataFrame(raw_data3, columns =['total_articulos'])
df3
```

```python
raw_data4 = re.findall('ATENDIDO POR :\s(.+?)\n' ,  output_string.getvalue(), re.M)
# raw_data4
```

```python
df4 = pd.DataFrame(raw_data4, columns =['atendido_por'])
df4
```
